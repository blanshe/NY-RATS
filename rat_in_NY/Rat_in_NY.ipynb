# Mapping Rat Sight Incidents in New York City 
This is a project that is looking at areas where rat sight in New York is likely to come from.
-> I am using a rat sight, a New York state open data-set created in 2010. 
->I will be mapping out areas (zip codes) where rat sight are likely to come from sorted by frequency
->I will point out zip code with the highest and lowest number of rat incidents in New York 
->I will create a donut chart of rats incidents reported in a Borough per year. 
->I will finally create a column where any person can type a zip code of his/her choice and get the incidents that has been reported in that area.


```python
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline

df = pd.read_csv("Rat_Sightings.csv")
df.head()
```

```python
df["Created Date"] = df["Created Date"].apply(lambda x : pd.to_datetime(x, format="%m/%d/%Y %H:%M:%S %p"))
```

```python
df.sort_values(['Created Date'],ascending=True)
```

```python
df["Borough"].unique()
```

```python
Borough = df[["Borough","Unique Key"]].groupby(["Borough"]).count()
Borough.to_csv('Borough.csv')
Borough
```

```python
Borough_year = df[["Borough","Unique Key"]].groupby([df["Created Date"].map(lambda x : x.year),"Borough"]).count()
Borough_year.to_csv('Borough_year.csv')
Borough_year
```
