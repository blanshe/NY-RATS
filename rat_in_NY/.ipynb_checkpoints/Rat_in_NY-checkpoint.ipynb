```python
import pandas as pd

df = pd.read_csv("Rat_Sightings.csv")
df.head()
```

```python
df["Created Date"] = df["Created Date"].apply(lambda x : pd.to_datetime(x, format="%m/%d/%Y %H:%M:%S %p"))
```

```python
len(df)
```

```python
columns = (list(df.columns))
print(columns)
print(len(columns))
```

```python
df.sort_values(['Created Date'],ascending=True)
```

```python
df["Borough"].unique()
```

```python
df[["Borough","Unique Key"]].groupby(["Borough"]).count()
```

```python
df[["Borough","Unique Key"]].groupby([df["Created Date"].map(lambda x : x.year),"Borough"]).count()
```

```python
table = df[["Borough","Unique Key"]].groupby([df["Created Date"].map(lambda x : x.year),"Borough"]).count().reset_index()
table.to_csv("table.csv")
```

```python
df[["Incident Zip","Unique Key"]].groupby(["Incident Zip"]).count().sort_values(['Unique Key'],ascending=False)
```
